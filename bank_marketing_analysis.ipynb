## 4. Veri Ön İşleme

Veri setini modelleme için hazırlıyoruz:
- Eksik değerlerin işlenmesi
- Outlier (aykırı değer) temizleme
- Kategorik değişkenlerin encoding edilmesi

```python
X_train, X_test, y_train, y_test, preprocessor = load_and_preprocess_data('bank-additional-full.csv')
```

## 5. Sınıflandırma Modelleri

Üç farklı sınıflandırma algoritması uyguluyoruz:
1. **Decision Tree**: Karar ağacı, yorumlanabilir model
2. **Random Forest**: Ensemble metodu, yüksek doğruluk
3. **Logistic Regression**: Hızlı ve etkili doğrusal model

```python
classifier = ClassificationModels(X_train, X_test, y_train, y_test)
classifier.train_and_evaluate_all()
```

```python
classification_results = classifier.get_results_dataframe()
print("\n Sınıflandırma Modelleri Sonuçları:")
classification_results
```

```python
classifier.plot_model_comparison()
classifier.plot_confusion_matrices()
```

```python
feature_names = preprocessor.get_feature_names()
rf_importance = classifier.get_feature_importance('Random Forest', feature_names)
```

## 6. Regresyon Modelleri

Görüşme süresini (duration) tahmin etmek için regresyon modelleri:
1. **Linear Regression**: Temel doğrusal regresyon
2. **Ridge Regression**: Regularization ile overfitting önleme

```python
preprocessor_reg = DataPreprocessor('bank-additional-full.csv')
preprocessor_reg.load_data()
preprocessor_reg.handle_unknown_values()
preprocessor_reg.remove_outliers()

duration = preprocessor_reg.df['duration'].copy()

preprocessor_reg.encode_categorical_variables()
X, y = preprocessor_reg.split_features_target()

from sklearn.model_selection import train_test_split
X_train_reg, X_test_reg, y_train_reg, y_test_reg, duration_train, duration_test = train_test_split(
    X, y, duration, test_size=0.2, random_state=42, stratify=y
)

X_train_reg_scaled, X_test_reg_scaled = preprocessor_reg.normalize_features(X_train_reg, X_test_reg)

print("Regresyon için veri hazırlandı!")
```

```python
regressor = RegressionModels(X_train_reg_scaled, X_test_reg_scaled, duration_train, duration_test)

regressor.train_and_evaluate_all()
```

```python
regression_results = regressor.get_results_dataframe()
print("\n Regresyon Modelleri Sonuçları:")
regression_results
```

```python
regressor.plot_model_comparison()
regressor.plot_predictions_vs_actual()
regressor.plot_residuals()
```

## 7. Kümeleme Analizi

Müşteri segmentasyonu için kümeleme algoritmaları:
1. **K-Means**: Hızlı ve etkili kümeleme
2. **Hierarchical Clustering**: Hiyerarşik yapı analizi

```python
X_combined = np.vstack([X_train, X_test])

clusterer = ClusteringModels(X_combined)

optimal_k_info = clusterer.train_all()
```

```python
clustering_results = clusterer.get_results_dataframe()
print("\n Kümeleme Modelleri Sonuçları:")
clustering_results
```

```python
clusterer.plot_elbow_silhouette(optimal_k_info)
clusterer.plot_dendrogram()
clusterer.plot_cluster_distribution()
clusterer.plot_cluster_comparison()
```

## 8. Kapsamlı Sonuç Analizi

Tüm modellerin karşılaştırmalı analizi ve sonuçların yorumlanması.

```python
analyzer = ResultsAnalyzer()
analyzer.set_classification_results(classification_results)
analyzer.set_regression_results(regression_results)
analyzer.set_clustering_results(clustering_results)

analyzer.print_all_results()
```
